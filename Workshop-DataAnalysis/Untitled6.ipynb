```python
from matplotlib import pyplot as plt
```

```python
x=[10,20,30,40]
y=[25,50,75,100]
plt.plot(x,y)
plt.show()
```

```python
x=[10,20,30,40]
y=[25,50,75,100]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('its a graph')
plt.show()
```

```python
#to print the mutliple line graphs
x=[10,20,30,40]
y1=[11,22,33,44]
y2=[24,30,56,78]
plt.plot(x,y1,label='y1 graph')
plt.plot(x,y2,label='y2 graph')
plt.title('multiline graph')
plt.legned(loc='upper right')
plt.show()
```

```python
#grid graphs
x=[1,2,3,4]
y=[10,20,30,40]
plt.plot(x,y,marker='s',ms=10,mfc='red',mec='y',color='cyan',ls='dashed',lw=6)
plt.grid(color='green',ls='dashed')
plt.show
```
